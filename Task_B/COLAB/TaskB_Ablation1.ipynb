```python
#library
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.applications as ap
```

```python
#mount file from google drive
from google.colab import drive
drive.mount('/content/drive')
#grab the data
img512 = np.load('/content/drive/MyDrive/Colab Notebooks/FINAL_DL/img512.npy')
label  = np.load('/content/drive/MyDrive/Colab Notebooks/FINAL_DL/label.npy')
```

```python
#prepare data
X512_train = img512
y512_train = label
y512_train = tf.keras.utils.to_categorical(y512_train, 4)
```

```python
#model ablation0

model_ab0 =  keras.Sequential()
model_ab0.add(keras.layers.Conv2D(input_shape=(512,512,3),filters = 64, kernel_size = (3,3),padding = 'same',activation = 'relu'))
#model_ab0.add(keras.layers.Conv2D(filters = 64, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))

model_ab0.add(keras.layers.Conv2D(filters = 128, kernel_size = (3,3),padding = 'same',activation = 'relu'))
#model_ab0.add(keras.layers.Conv2D(filters = 128, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))

model_ab0.add(keras.layers.Conv2D(filters = 256, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.Conv2D(filters = 256, kernel_size = (3,3),padding = 'same',activation = 'relu'))
#model_ab0.add(keras.layers.Conv2D(filters = 256, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))

model_ab0.add(keras.layers.Conv2D(filters = 512, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.Conv2D(filters = 512, kernel_size = (3,3),padding = 'same',activation = 'relu'))
#model_ab0.add(keras.layers.Conv2D(filters = 512, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))

model_ab0.add(keras.layers.Conv2D(filters = 512, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.Conv2D(filters = 512, kernel_size = (3,3),padding = 'same',activation = 'relu'))
#model_ab0.add(keras.layers.Conv2D(filters = 512, kernel_size = (3,3),padding = 'same',activation = 'relu'))
model_ab0.add(keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))

model_ab0.add(keras.layers.Flatten())
model_ab0.add(keras.layers.Dense(units=4096,activation='relu'))
model_ab0.add(keras.layers.Dense(units=4096,activation='relu'))
model_ab0.add(keras.layers.Dense(units=4,activation='softmax'))
```

```python
#VGG16
def model_vgg16(model,X_train, y_train,epochs = 8):
    model.compile(optimizer='AdaDelta', loss='categorical_crossentropy',metrics=['accuracy'])
    history = model.fit(X_train,y_train,batch_size=8, epochs=epochs, validation_split = 0.2)

    return model,history
```

```python
model_ab0,history0= model_vgg16(model_ab0,X512_train,y512_train,16)
```
