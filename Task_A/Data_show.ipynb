```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#use the plt inside the notebook
%matplotlib inline
```

```python
#grab the label
data = pd.read_pickle('./dataset/trainY.pkl')
```

```python
#to calculate the number of having tumor in samples
tumor_no = 0
tumor_meningioma = 0
tumor_glioma = 0
tumor_pituitary = 0

for i in range(3000):
    if data.tumor_type[i] == 'no':
        tumor_no = tumor_no + 1
    elif data.tumor_type[i] == 'meningioma':
        tumor_meningioma = tumor_meningioma + 1
    elif data.tumor_type[i] == 'glioma':
        tumor_glioma = tumor_glioma + 1
    elif data.tumor_type[i] == 'pituitary':
        tumor_pituitary = tumor_pituitary + 1

tumor_yes = [tumor_meningioma,tumor_glioma,tumor_pituitary]
tumor_data = [tumor_no,tumor_yes[0]+tumor_yes[1]+tumor_yes[2]]
```

```python
#prepare the color system
color1 = '#9a8f88'
color2 = '#aba199'
color3 = '#bbb2a8'
color4 = '#c9c3b6'
color5 = '#d5d1c3'
```

```python
#show the distribution
size = 0.33
fig, ax = plt.subplots(figsize = [6,6])
#outer
ax.pie([tumor_no]+tumor_yes,colors = [color1,color2,color3,color4],labels = ['no tumor','meningioma tumor','glioma tumor','pituitary tumor'],
       radius = 1,wedgeprops=dict(width=size, edgecolor='w'),autopct='%1.1f%%',pctdistance=0.85)
#inner
ax.pie(tumor_data,colors = [color1,color5],radius = 1-size,wedgeprops=dict(width=size, edgecolor='w'),autopct='%1.1f%%',pctdistance=0.75)

ax.legend(loc = 'upper right',bbox_to_anchor = (1.4, 1))
print(str(tumor_no)+' and '+str(tumor_yes))
```
